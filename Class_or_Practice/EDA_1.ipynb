```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
maths = [10,20,24,29,30]
english = [5,9,13,20,25]
age = [19,22,23,21,23]

plt.plot(maths,age,label="Maths",marker='o',ls='-.',lw=2,c='orange')
plt.plot(english,age,label='English',marker='x',ls='--', lw=2, c='magenta')

plt.xlabel("Marks")
plt.ylabel("Ages")

plt.legend(loc="best")

plt.annotate("Maths top",xy=(30,23))


plt.show()
```

```python
emp = ['Majid','Faheem','Amjid','Arslan','Danish']
sal = [10000,25000,40000,30000,35000]
bonus= [2000, 1400, 2300,1000,1500]

index = np.arange(len(emp))

fig, ax = plt.subplots(nrows=1,ncols=1)

ax.bar(x=index-0.15, height=sal, width=0.3, color='green')
ax.bar(x=index+0.15, height=bonus, width=0.3, color='orange')

#ax.set_xticks(ticks=[5000,10000,15000,20000,25000,30000,35000,40000])
ax.set_xticks(emp)

ax.set_xlabel("Employee salary/Bonus")
ax.set_ylabel("Employee Age")
```
