```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]
```

```python
fig , (ax1,ax2) = plt.subplots(nrows=2,ncols=1)
ax1.plot(years,chemical_exports,label="chemical",marker='o',ls=':',lw=2)
ax1.plot(years,medicine_exports,label='medicine',marker='x',ls='-.',lw=2)
ax2.plot(years,medicine_exports,label="medicine")

ax1.legend(loc='best')
ax2.legend(loc='best')

ax1.set_xlabel('Years')
ax1.set_ylabel('Medicine/Chemical')

xvals = np.arange(2009,2021,2)
#yvals = np.linspace(0.6,0.9,4)

#ax1.set_xticks([2010,2012,2015,2018,2022],labels=['awi','awi','awi','awi','awi'])
ax1.set_xticks(xvals)
#ax1.set_yticks()

ax1.annotate('top of chemicals',xy=(2016,0.95))

plt.tight_layout()


plt.show()
```
